```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("bezdekIris.data",  names=["sepal length", "sepal width", "petal length", "petal width", "type"])
df.info()
df.describe()

irises = df.drop(columns=["type"])
corr_df = irises.corr()
corr_df
```

```python
df["ratio_petal"] = (df["petal length"] / df["petal width"]).round(2)
df

```

```python
plt.figure(figsize=(5,5))
sns.heatmap(corr_df, annot=True,fmt=".2f", cmap="coolwarm")
plt.title("Correlation of Iris")
plt.show()
```

```python
sns.pairplot(df, hue="type")
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.boxplot(data=df, x="type", y="ratio_petal")
plt.title("Ratio petal by Iris Types")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x="ratio_petal", y="petal length")
plt.title("Ratio petal by petal length")
plt.show()
```
